```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from matplotlib.dates import DateFormatter

dates = np.arange(len(kk_counts2))
counts = kk_counts2.values

X = sm.add_constant(dates) 
model = sm.OLS(counts, X)
results = model.fit()

slope = results.params[1]
p_value = results.pvalues[1]

correlation_coefficient = np.corrcoef(dates, counts)[0, 1]

fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x=dates, y=counts, ax=ax, marker='o', linestyle='-', label='KK')
ax.plot(dates, results.predict(), color='red', linestyle='--', label='Linear Regression')

ax.set_xlabel('Time')
ax.set_ylabel('Comment Count')
ax.set_title('Distribution of KK Mentions with Linear Regression')

date_labels = kk_counts2.index.strftime('%b %Y')
first_day_indices = np.where(kk_counts2.index.day == 1)[0]
first_day_ticks = dates[first_day_indices]
ax.set_xticks(first_day_ticks)
ax.set_xticklabels(date_labels[first_day_indices])

ax.legend()
text = f'Slope: {slope:.2f}\n'
text += f'p-value: {p_value:.4f}\n'
text += f'Correlation Coefficient: {correlation_coefficient:.2f}'
ax.text(0.7, 0.9, text, transform=ax.transAxes, bbox=dict(facecolor='white', edgecolor='gray', boxstyle='round'))

plt.show()
```

```python
# general average
averages = df.resample('1D', on='publishDate')['sentiment_score'].mean() # split into 15 day periods

# plot graph
plt.figure(figsize=(14, 5))
averages.plot(kind='line', marker='o', linestyle='-')
plt.xlabel('Date')
plt.ylabel('Average Score')
plt.title('Average Score day by day period')
plt.ylim(0, 5)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy import stats

x = averages.index
y = averages.values
slope, intercept, r_value, p_value, std_err = stats.linregress(range(len(x)), y)

correlation_coefficient = r_value

plt.figure(figsize=(12, 5))
plt.plot(x, y, marker='o', linestyle='-')

regression_line = slope * range(len(x)) + intercept
plt.plot(x, regression_line, color='red', linestyle='--', label='Linear Regression')

plt.text(0.05, 0.9, f"Slope: {slope:.3f}", transform=plt.gca().transAxes)
plt.text(0.05, 0.85, f"P-value: {p_value:.3f}", transform=plt.gca().transAxes)
plt.text(0.05, 0.8, f"Correlation Coefficient: {correlation_coefficient:.3f}", transform=plt.gca().transAxes)

plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

plt.xlabel('Date')
plt.ylabel('Average Score')
plt.title('Average Score day by day period')
plt.ylim(0, 5)

plt.legend()

plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
