This part creates dataframes

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create dataframe
df=pd.read_csv("DataAnalysis.csv")
```

```python
# convert date information to datetime object
from datetime import datetime
df["publishDate"] = pd.to_datetime(df["publishDate"])
```

```python
# seperate dataframes that only keep comments that mention a specific presidential candidate
kk_df = df[df['KKscore'] != 0] 
rte_df = df[df['RTEscore'] != 0]
so_df = df[df['SOscore'] != 0] 
mi_df = df[df['MIscore'] != 0] 
```

In this part the average sentiment scores of each presidential candidate for each channel are displayed
