```python
# 15 gunluk chunklara ayirma islemi

# Convert the 'date' column to datetime if needed
sorted_df['publishDate'] = pd.to_datetime(sorted_df['publishDate'])

# Define the start and end dates of the overall date range
start_date = pd.to_datetime('2023-02-01')
end_date = pd.to_datetime('2023-05-27')

# Define the chunk size (number of days)
chunk_size = 15

# Iterate over the date range in 15-day chunks
date_range = pd.date_range(start=start_date, end=end_date, freq=f'{chunk_size}D')

# List to store the chunk DataFrames
chunk_dfs = []

for i in range(len(date_range) - 1):
    # Get the start and end dates for the current chunk
    chunk_start_date = date_range[i]
    chunk_end_date = date_range[i + 1] - pd.DateOffset(days=1)

    # Filter the DataFrame for the current chunk
    chunk_df = sorted_df[(sorted_df['publishDate'] >= chunk_start_date) & (sorted_df['publishDate'] <= chunk_end_date)]

    # Append the chunk DataFrame to the list
    chunk_dfs.append(chunk_df)
```

```python
# listenin elemanlarina tek tek erismek gerekiyor
first_chunk = chunk_dfs[0]
second_chunk = chunk_dfs[1]
```
